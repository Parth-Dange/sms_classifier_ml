## data preprosessing
1. lowercase 
2. tokenisation and removing special characters
3. removing stop words and punctuations
4. stemming


```python
import nltk
import string
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
def transform_text(text):
    text =text.lower()
    text = nltk.word_tokenize(text)
    
    y = []
    for i in text:
        if i.isalnum():
            y.append(i)
            
    text = y[:]  # cloning is very important to save the list into string format
    y.clear()
    
    for i in text:
        if i not in stopwords.words('english') and i not in string.punctuation:
            y.append(i)
            
    text = y[:]  # cloning is very important to save the list into string format
    y.clear()
    
    for i in text:
        y.append(ps.stem(i))
    
        
    return " ".join(y)
```

```python
transform_text('lectures are interesting but you are not $$')
```

```python
df['transfromed_text']  = df['text'].apply(transform_text)
```
